```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
def audio_to_spectrogram(audio, sr=16000, n_fft=8000, hop_length=256):
    stft_matrix = librosa.stft(audio, n_fft=n_fft, hop_length=hop_length)
    magnitude = np.abs(stft_matrix)  # 取幅度谱
    phase = np.angle(stft_matrix)  # 取相位谱
    return magnitude, stft_matrix.shape, phase


def plot_spectrogram(spectrogram, save_path=None):
    plt.figure(figsize=(10, 6))
    plt.imshow(spectrogram, aspect='auto', origin='lower', cmap='inferno')
    plt.colorbar(label="Amplitude")
    plt.title("Spectrogram Visualization")
    plt.xlabel("Time (frames)")
    plt.ylabel("Frequency (bins)")
    if save_path:
        plt.savefig(save_path, dpi=300)
        print(f"频谱图已保存至 {save_path}")
    plt.show()
```

```python
Audio('新年快乐.wav', rate=16000)
```

```python
audio, sr = librosa.load('blended_audio.wav', sr=16000)
spec, _, phase = audio_to_spectrogram(audio)
print(spec.shape)

# 绘制并保存频谱图
plot_spectrogram(spec, save_path='spectrogram_visualization.png')
```
